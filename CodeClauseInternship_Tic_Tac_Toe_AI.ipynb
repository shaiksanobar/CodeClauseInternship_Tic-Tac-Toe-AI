```python
# TIC-TAC-TOE AI - COMPLETE IN 5 MINUTES
import random

def print_board(board):
    print("\n")
    for i in range(3):
        print(f" {board[i*3]} | {board[i*3+1]} | {board[i*3+2]} ")
        if i < 2:
            print("-----------")

def check_winner(board):
    # Winning combinations
    lines = [(0,1,2), (3,4,5), (6,7,8),  # rows
             (0,3,6), (1,4,7), (2,5,8),  # columns
             (0,4,8), (2,4,6)]           # diagonals

    for a,b,c in lines:
        if board[a] == board[b] == board[c] != ' ':
            return board[a]
    return None

def is_board_full(board):
    return ' ' not in board

def ai_move(board):
    # Simple AI: Try to win, then block, then random
    # 1. Try to win
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            if check_winner(board) == 'O':
                return i
            board[i] = ' '

    # 2. Block player
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'X'
            if check_winner(board) == 'X':
                board[i] = 'O'
                return i
            board[i] = ' '

    # 3. Random move
    empty_spots = [i for i in range(9) if board[i] == ' ']
    return random.choice(empty_spots)

def play_game():
    board = [' '] * 9
    print("ðŸŽ® TIC-TAC-TOE AI")
    print("You are X, AI is O")
    print("Enter position (1-9):")
    print(" 1 | 2 | 3 ")
    print("-----------")
    print(" 4 | 5 | 6 ")
    print("-----------")
    print(" 7 | 8 | 9 ")

    while True:
        print_board(board)

        # Player move
        try:
            pos = int(input("\nYour move (1-9): ")) - 1
            if board[pos] != ' ':
                print("Spot taken! Try again.")
                continue
            board[pos] = 'X'
        except:
            print("Invalid input! Enter 1-9.")
            continue

        # Check if player won
        if check_winner(board) == 'X':
            print_board(board)
            print("ðŸŽ‰ YOU WIN!")
            break

        if is_board_full(board):
            print_board(board)
            print("ðŸ˜ IT'S A TIE!")
            break

        # AI move
        print("\nðŸ¤– AI is thinking...")
        ai_pos = ai_move(board)
        board[ai_pos] = 'O'

        # Check if AI won
        if check_winner(board) == 'O':
            print_board(board)
            print("ðŸ¤– AI WINS!")
            break

        if is_board_full(board):
            print_board(board)
            print("ðŸ˜ IT'S A TIE!")
            break

# START THE GAME
play_game()
```

```python
# EXTRA: PLAY AGAIN FEATURE
def play_again():
    while True:
        play_game()
        again = input("\nPlay again? (y/n): ").lower()
        if again != 'y':
            print("Thanks for playing! ðŸ‘‹")
            break

play_again()
```
